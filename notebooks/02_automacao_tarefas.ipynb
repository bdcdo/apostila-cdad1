Existem regras básicas sobre como uma função é definida. 

1) A linha inicia com a palavra 'def';

2) Logo depois há um espaço e após o espaço o nome escolhido para a função;

3) Após o nome da função abre-se parênteses e o argumento ou os argumentos da função são definidos;

4) Fechados os parênteses, necessariamente há ':';

5) Após ':', inicia-se uma nova linha com indentação. Todas as linhas seguintes com indentação estão dentro do escopo da função. Isso significa que tais comandos serão executados sempre que a função for chamada;

6) A função pode devolver um ou mais objetos. Esse comando será definido pela linha que inicia com 'return'. Após a linha que inicia com 'return', as linhas seguintes não serão executadas.

Quando analisamos alguns tipos penais do Código Penal foi possível perceber que existem condições especiais, para cada crime, que causam aumento ou diminuição da pena. No caso do homicídio, existem condições que aumentam a pena e condições que diminuem a pena.

Considere os seguintes casos concretos:

Will matou Chris, a pena é de 9 anos.

Maria matou Daniela, a pena é de 8 anos. O crime foi cometido por relevante valor social, sob domínio de violenta emoção, logo após injusta provocação de Daniela. A pena é diminuída em 1/6. 

Jorge matou Mário, a pena é de 7 anos. O crime foi cometido por milícia privada, sob pretexto de prestar serviço de segurança. A pena é aumentada em 1/3.

Pedro matou Gabriela em contexto de violência doméstica, a pena é de 15 anos. O crime foi cometido quando Gabriela estava grávida, a pena é aumentada em 1/2.  

Há um padrão na aplicação da pena: em todos os casos existe uma pena base que é aumentada ou diminuída segundo algum dispositivo do art. 121. Vamos construir uma função que realize automaticamente o aumento nos casos de crime praticado por milícia privada sob pretexto de prestação de serviço de segurança. Nesses casos o aumento é de 1/3 até 1/2. Nossa função ira tomar uma pena base e realizar o aumento mínimo de 1/3.


```python
def aumento_milicia(pena):
  pena_aumentada = pena + pena/3
  return pena_aumentada

aumento_milicia(10)
```

Uma vez que a função tenha sido definida, podemos chamá-la com diferentes valores de pena base. Basta passar o valor da pena base como argumento entre ().

Funções podem conter comandos muito mais sofisticados. Os comandos no escopo das regras da função podem ter 20 ou até 100 linhas de instruções. Podem conter condicionais e inclusive iterações.

Vamos montar uma função que emula o trabalho de dosimetria da pena feito pela juíza na sentença. A função recebe uma lista com crimes, checa se cada crime tem condições especiais para aumento ou diminuição de pena e devolve o dicionário com as penas calculadas.

```python
def dosimetria(crimes):
  for crime in crimes:
    if 'valor social' in crime['circunstâncias']:
      crime['pena final'] = crime['pena base'] - crime['pena base']/6
    if 'milícia' in crime['circunstâncias']:
      crime['pena final'] = crime['pena base'] + crime['pena base']/3
  return crimes


crimes = [{'pena base':9, 'circunstâncias':''},
          {'pena base':8, 'circunstâncias':'O crime foi cometido por relevante valor social, sob domínio de violenta emoção, logo após injusta provocação de Daniela.'},
          {'pena base':7, 'circunstâncias':'O crime foi cometido por milícia privada, sob pretexto de prestar serviço de segurança.'}]

dosimetria(crimes)    
```

Quando o objetivo da função é muito simples - a ponto de todas as regras da função caberem em uma única expressão - é possível adotar o formado simplificado de funções. A definição segue regras mais fáceis:

O nome dessas funções simplificadas é lambda. Elas podem receber vários argumentos, se necessário:

Em muitas situações será mais fácil, eficiente ou elegante definir e usar uma função lambda do que uma função normal.

## Map <a class="anchor" id="second-bullet"></a>

Frequentemente será necessário realizar repetidamente um mesmo conjunto de comandos para vários itens de um iterável. Se esses comandos já estão definidos em uma função, não há necessidade de iterar usando 'for' por cada item. A função nativa 'map()' recebe dois argumentos. O primeiro é uma função de sua escolha. O segundo é o objeto iterável. A função 'map()' então retornará um iterável contendo o resultado da chamada da função para cada item do iterável.

```python
def aumento_milicia(pena):
  pena_aumentada = pena + pena/3
  return pena_aumentada

penas_originais = (6, 6, 8, 9, 10, 10, 10)

penas_calculadas = map(aumento_milicia,penas_originais)
print(list(penas_calculadas))
```

A função 'map()' pode ser usada para editar muitas strings de forma padronizada.

```python
def separa_art(texto):
  texto = texto.split('.')
  return texto[1]

artigos = ['Art. 79. São bens imóveis o solo e tudo quanto se lhe incorporar natural ou artificialmente.',
           'Art. 80. Consideram-se imóveis para os efeitos legais: I - os direitos reais sobre imóveis e as ações que os asseguram; II - o direito à sucessão aberta.',
           'Art. 81. Não perdem o caráter de imóveis: I - as edificações que, separadas do solo, mas conservando a sua unidade, forem removidas para outro local; II - os materiais provisoriamente separados de um prédio, para nele se reempregarem.']

somente_numeros = map(separa_art,artigos)
print(tuple(somente_numeros))
```

Note que na função 'map()' o primeiro argumento é justamente uma função. Ela já foi previamente definida ou pode ser definida na própria chamada de 'map()' usando o formato lambda.

```python
penas_originais = (6, 6, 8, 9, 10, 10, 10)
penas_calculadas = map(lambda x: x+x/3, penas_originais)
print(list(penas_calculadas))
```

Ao definir as regras da função lambda dentro da chamada de penas_calculadas não foi sequer necessário dar um nome à função que calcula o aumento de pena. É por essa razão que funções lambda são também conhecidas como funções anônimas.

## Zip <a class="anchor" id="third-bullet"></a>

A função 'zip()' é similar à 'map()' porque também otimiza algo repetitivo. Ela permite associar, em ordem, diferentes grupos de objetos.

```python
penas = (6,12,0.25,4)
crimes = ('homicídio','feminicídio','lesão corporal','roubo')

agrupados = zip(penas,crimes)
print(list(agrupados))
```

Cada novo conjunto gerado é uma *tupla* e possui um item do primeiro iterável passado (penas) e um item do segundo iterável passado (crimes). 

Por exemplo, a terceira tupla possui o terceiro item da tupla 'penas' e o terceiro item da tupla 'crimes'.

A função 'zip()' permite agrupar pares de objetos, trios, quartetos e assim por diante, sempre em tuplas.

```python
pena_minima = (6,12,0.25,4)
pena_maxima = (20,30,1,10)
crime = ('homicídio','feminicídio','lesão corporal','roubo')

infos = zip(pena_minima, pena_maxima, crime)

def imprimir_infos(trio):
  return f"A pena mínima para {trio[2]} é {trio[0]} anos e a pena máxima é {trio[1]} anos."

print(list(map(imprimir_infos,infos)))
```

Se 'zip()' receber listas ou tuplas com extensões diferentes não haverá erro na execução do código. Mas o objeto com os agrupamentos somente possuirá tantos agrupamentos quanto o número de itens na menor lista ou tupla passada.

```python
pena_minima = (6,12,0.25,4)
pena_maxima = (20,30,1,10)
crime = ('homicídio','feminicídio','lesão corporal')

infos = zip(pena_minima, pena_maxima, crime)
print(list(infos))

print(len(pena_minima))
print(len(list(infos)))
```
