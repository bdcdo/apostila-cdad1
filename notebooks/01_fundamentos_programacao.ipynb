Abstrações são uma ferramenta importante do Direito e da Ciẽncia da Computação. No Direito, a abstração é uma característica das normas jurídicas que permite separar a norma em si da sua aplicação a um caso concreto ou suporte fático. Isso significa que a aplicação da norma a um caso específico não altera ou esgota a própria norma, sua validade ou eficácia.

O artigo 12 do Código de Defesa do Consumidor estabelece a responsabilidade civil do "fornecedor" pelo fato do produto ou serviço.

**"Art. 12. O fabricante, o produtor, o construtor, nacional ou estrangeiro, e o importador respondem, independentemente da existência de culpa, pela reparação dos danos causados aos consumidores por defeitos decorrentes de projeto, fabricação, construção, montagem, fórmulas, manipulação, apresentação ou acondicionamento de seus produtos, bem como por informações insuficientes ou inadequadas sobre sua utilização e riscos."**

Se Maria adquire um veículo da fabricante (fornecedora) Toyota e um defeito desse veículo vem a causar um dano a Maria - uma lesão corporal, por exemplo, então a aplicação do artigo 12 do CDC resulta em uma obrigação, no caso específico, que a Toyota tem de reparar o dano de Maria. Uma forma possível de reparação do dano é a indenização.


A norma do CDC em abstrato pode ser expressada da seguinte forma:

```python
art_12_abstrato = "Art. 12. O fabricante, o produtor, o construtor, nacional ou estrangeiro, e o importador respondem, independentemente da existência de culpa, pela reparação dos danos causados aos consumidores por defeitos decorrentes de projeto, fabricação, construção, montagem, fórmulas, manipulação, apresentação ou acondicionamento de seus produtos, bem como por informações insuficientes ou inadequadas sobre sua utilização e riscos."
```

A principal desvantagem da tupla é que uma vez criada ela não pode ser modificada para adicionar ou remover um dos objetos do conjunto.

```python
dados_processo + 'novo item'
```

## Condicionais e dicionários <a class="anchor" id="second-bullet"></a>

O Código Penal estipula tipos seguindo a necessidade de abstração e impessoalidade do Direito. O crime de homicídio simples é definido no seguinte formato:

*"Art. 121. Matar alguem: Pena - reclusão, de seis a vinte anos."*

A incidência do tipo penal a uma conduta praticada contra vítima específica pode ser representada da seguinte maneira:

```python
alguem = 'Odete Roitman'

art_121_CP = f'Matar {alguem}: Pena - reclusão, de seis a vinte anos.'

print(art_121_CP)
```

Mas não sabemos *se* ela realmente foi vítima do crime. 

Responder *se* um fato ocorreu (suporte fático) e *se* uma norma pode ser aplicada a esse fato é a essência da atividade de qualquer operadora do Direito. A ordem lógica do raciocínio jurídico pode ser expressada da seguinte maneira em português:


**Se o fato ocorreu:**

**Aplica-se a norma.**

Temos uma condição estipulada e um resultado a implementar caso essa condição seja preenchida.

É possível fazer a mesma formulação em Python:

```python
if 'fato' == 'ocorreu':
  print('Aplicar a norma ao caso.')
```

O sinal 

== 

é fundamentalmente diferente do sinal 

= 

em Python. Enquanto = expressa um comando, uma ordem - a partir desse momento, essa variável está atrelada a esse objeto - o sinal == é uma pergunta, um teste. Essa pergunta é feita à máquina, que recebe instruções sobre o que fazer caso a condição seja preenchida.

```python
odete_roitman = 'vítima'
if odete_roitman == 'vítima':
    print('O crime foi praticado.')
print('mas e essa linha?')
```

A condição pode ser formulada em sentido negativo. Pergunta se Odete Roitman é a vítima é o oposto de perguntar se ela não é a vítima.

```python
if odete_roitman != 'vítima':
    print('O crime não foi praticado.')
```

Condições podem ser expressadas utilizando o operador de igualdade, o de desigualdade e operadores de maior ou menor.

```python
pena_homicidio_simples = 6
pena_homicidio_qualificado = 12

if pena_homicidio_simples < pena_homicidio_qualificado:
    print('A pena do art. 121, caput é inferior à pena do art. 121, §2 do Código Penal.')

pena_furto = 1

if pena_homicidio_simples > pena_furto:
    print('A pena do art. 121, caput é superior à pena do art. 155, caput do Código Penal.')

pena_homicidio_culposo = 1
```

É possível expressar uma condição mais ampla do que apenas igualdade ou superioridade. O sinal >= significa que a condição estará preenchida em ambas as hipóteses.

```python
if pena_homicidio_culposo >= pena_furto:
    print('A pena do art. 155 é superior ou igual à pena do art. 121, §3 do Código Penal.')
```

Mas o que acontece quando a condição não é preenchida? Nada. Se necessário, podemos especificar uma espécia de situação residual. Ou seja, o que deve ser implementado na hipótese de a condição não ser satisfeita:

```python
if pena_homicidio_culposo == pena_furto:
    print('A pena do art. 155 é igual à pena do art. 121, §3 do Código Penal.')
else:
    print('A pena do art. 155 não é igual à pena do art. 121, §3 do Código Penal.')
```

Como ambos tipos penais têm como pena mínima 1 ano, a condição foi preenchida e a hipótese residual não foi acionada. O else não expressa uma condição propriamente. Note que entre 'else' e ':' não há qualquer regra estipulada. Dessa forma, não importa o que aconteça, se a condição imediatamente anterior não foi preenchida, o que estiver estipulado no else será implementado.

**Múltiplas Condições**

É possível estabelecer múltiplas condições permitindo que mais de uma delas - ou todas - sejam aceitas. O crime de homicídio qualificado, previsto no art. 121, §2 do Código Penal, é caracterizado por ser mais grave do que o homocídio simples. A conduta é enquadrada no tipo penal do homicídio qualificado se o existe ao menos uma das características previstas nos 5 incisos. Não é necessário que a conduda preencha a condição de 2 ou 3 dos incisos. Apenas um deles é suficiente.


```python
crime_cometido = ['por motivo fútil','para assegurar a execução, a ocultação, a impunidade ou vantagem de outro crime']
```

```python
#inciso I
if 'mediante paga ou promessa de recompensa, ou por outro motivo torpe' in crime_cometido:
    print('Foi praticado homicídio qualificado por recompensa ou promessa de recompensa, ou por outro motivo torpe.')
#inciso II
if 'por motivo fútil' in crime_cometido:
    print('Foi praticado homicídio qualificado por motivo futil.')
#inciso III
if 'com emprego de veneno, fogo, explosivo, asfixia, tortura ou outro meio insidioso ou cruel, ou de que possa resultar perigo comum' in crime_cometido:
    print('Foi praticado homicídio qualificado com emprego de veneno, fogo, explosivo, asfixia, tortura ou outro meio insidioso ou cruel, ou de que possa resultar perigo comum.')
#inciso IV
if 'à traição, de emboscada, ou mediante dissimulação ou outro recurso que dificulte ou torne impossivel a defesa do ofendido' in crime_cometido:
    print('Foi praticado homicídio qualificado à traição, de emboscada, ou mediante dissimulação ou outro recurso que dificulte ou torne impossivel a defesa do ofendido.')
#inciso V  
if 'para assegurar a execução, a ocultação, a impunidade ou vantagem de outro crime' in crime_cometido:
    print('Foi praticado homicídio qualificado para assegurar a execução, a ocultação, a impunidade ou vantagem de outro crime.')
```

Havia 5 condições e tanto a 2a quanto a 5a foram satisfeitas. Isso signifca que o homicídio não é simples, mas sim qualificado. Sob o ponto de vista jurídico, uma representação mais precisa seria: 

```python
#inciso I
if 'mediante paga ou promessa de recompensa, ou por outro motivo torpe' in crime_cometido:
    print('Foi praticado homicídio qualificado.')
#inciso II
elif 'por motivo fútil' in crime_cometido:
    print('Foi praticado homicídio qualificado.')
#inciso III
elif 'com emprego de veneno, fogo, explosivo, asfixia, tortura ou outro meio insidioso ou cruel, ou de que possa resultar perigo comum' in crime_cometido:
    print('Foi praticado homicídio qualificado.')
#inciso IV
elif 'à traição, de emboscada, ou mediante dissimulação ou outro recurso que dificulte ou torne impossivel a defesa do ofendido' in crime_cometido:
    print('Foi praticado homicídio qualificado.')
#inciso V  
#elif 'para assegurar a execução, a ocultação, a impunidade ou vantagem de outro crime' in crime_cometido:
  #print('Foi praticado homicídio qualificado.')
else:
    print('Foi praticado homicídio simples.')
```

Caso nenhuma das 5 condições fosse satisfeita, a sentença deixaria claro que o homicídio é do tipo simples.

Nesse caso, existe uma maneira ainda mais eficiente de testar as mesmas condições e emitir uma sentença:

```python
if 'mediante paga ou promessa de recompensa, ou por outro motivo torpe' or 'por motivo fútil' or 'com emprego de veneno, fogo, explosivo, asfixia, tortura ou outro meio insidioso ou cruel, ou de que possa resultar perigo comum' or 'à traição, de emboscada, ou mediante dissimulação ou outro recurso que dificulte ou torne impossivel a defesa do ofendido' or 'para assegurar a execução, a ocultação, a impunidade ou vantagem de outro crime' in crime_cometido:
    print('Foi praticado homicídio qualificado.')
else: 
    print('Foi praticado homicídio simples.')  
```

No Direito e aplicação do Código Penal as condições podem ser alternativas, como acima, ou cumulativas. Quando as condições são cumulativas, é necessário que todas elas estejam igualmente satisfeitas. O tipo penal de roubo elenca condições cumulativas e alternativas:

*"Art. 157 - Subtrair coisa móvel alheia, para si ou para outrem, mediante grave ameaça ou violência a pessoa, ou depois de havê-la, por qualquer meio, reduzido à impossibilidade de resistência"*

Existe uma condição "Subtrair coisa móvel alheia, para si ou para outrem" que é necessária. Se apenas ela fosse preenchida, o crime seria de furto e não roubo. Para que a conduta seja enquadrada como roubo é necessário que ao menos uma de outras três condições esteja preenchida juntamente com a subtração de coisa alheia móvel. Que essa subtração seja 

1) mediante grave ameaça a pessoa ou 

2) mediante violência a pessoa ou 

3) após ter reduzido a vítima à impossibilidade de resistência. A norma jurídica do art. 157 do Código Penal pode ser expressada da seguinte maneira em Python:


```python
#crime_cometido = ['subtração de coisa alheia móvel']
crime_cometido = ['subtração de coisa alheia móvel', 'mediante grave ameaça a pessoa']

if 'subtração de coisa alheia móvel' in crime_cometido and ('mediante grave ameaça a pessoa' or 'mediante violência a pessoa' or 'após ter reduzido a vítima à impossibilidade de resistência') in crime_cometido:
    print('Foi praticado crime de roubo.')
```

A implementação da condição não precisa ficar limitada a um único comando e esse comando não precisa ser a chamada da função print().

## Iteração e Range <a class="anchor" id="fourth-bullet"></a>


Um dos elementos mais decisivos da automação - em qualquer mercado ou indústria - é a otimização de tarefas repetitivas. No mercado jurídico isso não é diferente. Varas lidam com centenas ou milhares de processos similares ao longo do ano e com muitos milhares de despachos formais que não exigem uma decisão subjetiva. Escritórios possuem carteiras de milhares ou dezenas de milhares de processos com similaridades entre si e que tramitam em uma mesma comarca e/ou possuem objetos muito parecidos. Por exemplo, muitas ações judiciais cujo objeto é demissão sem pagamento das obrigações trabalhistas. 
Uma boa jurista deve saber identificar os elementos mais ou menos otimizáveis de um conjunto de tarefas repetitivas para poder adequadamente separar aquilo que precisa ser feito e avaliado por juristas, de um lado, daquilo que pode e deve ser delegado para a automatização, de outro.

*Fonte: Casengine.app*

Alguns problemas jurídicos não podem ser automatizados, outros podem. Para aqueles que podem ser automatizados, é preciso identificar a parte da tarefa que é mais repetitiva e padronizada. A partir daí, um padrão ou regra geral pode ser formulado para elaborar comandos que serão executados pela máquina.
Isso quase sempre envolve um círculo de repetição de um conjunto de comandos para cada processo judicial, para cada contrato, para cada cliente etc. 
Esse círculo de repetição na ciência da computação é chamado de **iteração**. Determinada tarefa jurídica pode precisar ser repetida para 100 processos, o que significa 100 iterações de um mesmo conjunto de comandos. Outra tarefa pode precisar ser repetida para 1000 despachos em uma vara, o que significa 1000 iterações de outro conjunto de comandos.

*Fonte: flaticon.com*

Ao estabelecer os comandos de algo que irá ser executado repetidas vezes, a juristas pode estar em uma de duas situações:

1) **Sabe de antemão o número exato de vezes que os comandos serão repetidos**. Se possui uma lista de 250 petições iniciais e para cada petição é precisa checar se os argumentos possuem fundamentação jurídica, então a jurista já sabe que a tarefa será repetida exatamente 250 vezes.

2) **Não sabe de antemão o número exato de vezes que os comandos serão repetidos**. Se a jurista está tentando mediar um acordo entre a empresa cliente e um consumidor insatisfeito, não há como saber se o consumidor irá aceitar a primeira, a segunda, a terceira, a quarta oferta e assim por diante. Haverá um conjunto de tarefas padronizadas para cada nova oferta, mas essas tarefas podem ser executadas uma única fez, se o consumidor aceitar a proposta inicial, ou sete vezes, se o consumidor achar que as primeiras propostas não foram boas o suficiente.

Vamos ver alguns exemplos do primeiro tipo de iteração. O formato será sempre o seguinte:

**for** objeto **in** objeto_iterável**:**

Na nova linha deverá haver uma indentação. Todas as linhas seguintes que também possuírem a **indentação** serão consideradas como **parte do escopo** do for loop. A primeira linha sem a indentação marcará o final do escopo do for. 
**Os comandos que serão repetidos são exclusivamente aqueles dentro do escopo do for.** 

```python
palavras = ('Penal', 'Constitucional', 'Processual', 'Contratual')
tamanhos = {}
for i in palavras:
    print('Nessa iteração, a variável palavra está associada ao objeto',i,', que é do tipo',type(i))
    tamanhos[i] = len(i)
    print(i)

print(i)
print(tamanhos)
```

O formato da iteração com **while** é similar àquela com for em pelo menos dois sentidos:
1) ao final da linha haverá necessariamente : e na nova linha uma indentação. Todas as linhas seguintes que possuírem indentação serão parte do escopo do while.
2) Todas as linhas/comandos parte do escopo do while serão repetidas a cada iteração.

O formato da iteração com while é diferente daquela com for em pelo menos dois sentidos:
1) Não há um objeto sobre o qual haverá uma iteração.
2) O número de iterações será definido apenas pela condição estabelecida logo após o while. No início de cada nova iteração a condição é testada. Se ela é satisfeita, a nova iteração é executada. Se a condição não é satisfeita, não haverá mais novas iterações.


```python
total = 15
adicao = 0
iteracao = 0
while adicao < total:
    iteracao += 1
    print('Na iteração',iteracao,'o valor de adicao é',adicao)
    adicao += 2
print('Após executadas as iterações, o valor de adicao é',adicao)
```

Nem sempre é possível prever com precisão o número de vezes que o mesmo conjunto de comandos precisará ser repetido. Quando se trata de mediação, não há como saber o número de ofertas feitas à outra parte até que seja possível chegar a um consenso.

```python
tentativa = 2200
while tentativa < 3000:
    tentativa += 150
    resposta = input('Gostaria de fechar acordo por '+str(tentativa)+ '?')
    if resposta == 'Sim':
        break
    if resposta == 'Não':
        print('Tudo bem, formularemos uma nova proposta.')

if resposta == 'Sim':
    print('Ótimo! Nossa equipe jurídica irá entrar em contato para assinar o acordo no valor de '+str(tentativa))
else:
    print('Infelizmente não temos condições de oferecer um valor mais alto do que '+str(tentativa))
```

**Muita atenção!** Um loop em for é necessariamente finito, pois o objeto sobre o qual se está iterando (string, lista, tupla, range etc) possui um número finito de partes.

Um **loop em while pode ser infinito** se a condição estabelecida sempre for preenchida. Note no exemplo acima duas maneiras de encerrar o while loop:
1) O valor de 'tentativa' é alterado a cada iteração na primeira linha do escopo do while. Isso significa que em algum momento 'tentativa' terá valor maior do que a média do valor da última proposta de acordo em nossa base. Quando isso acontecer, o while loop irá encerrar.
2) A palavra-chave **'break'** foi inserida como comando a ser executado se a qualquer momento a resposta do consumidor para a pergunta for 'Sim'. O comando break força o encerramento de qualquer loop ou iteração, em for ou while. No caso do while, o break força o final do while loop independentemente da satisfação ou não da condição do while. No caso do for loop, o comando break força o final da iteração mesmo que a iteração não tenha percorrido todos os itens do objeto iterado.

```python
palavras = ('Penal', 'Constitucional', 'Processual', 'Contratual')
tamanhos = {}
for palavra in palavras:
    if palavra == 'Processual':
        break
    print('Nessa iteração, a variável palavra está associada ao objeto',palavra,', que é do tipo',type(palavra))
    tamanhos[palavra] = len(palavra)

print(tamanhos)
```
